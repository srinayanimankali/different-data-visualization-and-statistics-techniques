```python
#  According to null hypothesis the mean value of the funds with operating state as operating must be equal to closed i.e H0= M1=M2. The alternative hypothesis is the opposite of the null hypothesis ie. H1: M1!=M2. ie Both the means are different from each other

```

```python
#Ztest and sample t test are used to compare the values and decide whether there is significant difference between the funds of Operating state closed and operating

from statsmodels.stats.weightstats import ztest
op1= df12.loc[df12.OperatingState == 'Operating', 'funding in Millions(M)']
op2 = df12.loc[df12.OperatingState =='Closed', 'funding in Millions(M)']
alpha = 0.05   # Let's consider a significance level of 5%
test_statistic, p_value = ztest(op1, op2)  # Z test

if p_value <= alpha:
    print(f'Since the p-value, {round(p_value, 3)} < {alpha} (alpha) the difference is significant so, reject the Null hypothesis')
else:
    print(f'''\t Since the p-value, {round(p_value,3)} > {alpha} (alpha) the difference is not significant and,
    \t so we fail to reject the Null hypothesis''')
```

```python
# sample t test
from scipy.stats import ttest_ind

test_statistic, p_value = ttest_ind(op1, op2)

if p_value <= alpha:
    print(f'Since the p-value, {round(p_value, 3)} < {alpha} (alpha) the difference is significant and we reject the Null hypothesis')
else:
    print(f'''\t Since the p-value, {round(p_value,3)} > {alpha} (alpha) the difference is not significant and,
    \t we fail to reject the Null hypothesis''')
```

Therefore, according to the z test and t tests there is no significant difference so, we fail to reject the null hypothesis.
